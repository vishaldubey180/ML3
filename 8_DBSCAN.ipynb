### DBSCAN on Density dataset 

```python
import pandas as pd

df = pd.read_csv("datasets/Density.csv")

df = df.iloc[:, 1:]

df = df.rename(columns={"0": "a", "1": "b"})

df.sample(5)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["a"], df["b"])
```

```python
from sklearn.cluster import DBSCAN

model = DBSCAN(eps=30, min_samples=5)

model.fit(df)
```

```python
labels = model.labels_

plt.scatter(df["a"], df["b"], c=labels)
```
