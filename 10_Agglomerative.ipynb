### Agglomerative Clustering using mall_customer dataset

```python
import pandas as pd

df = pd.read_csv("datasets/Mall_Customers.csv")

df.sample(5)
```

```python
df = df.rename(columns={"Annual Income (k$)": "annual_income", "Spending Score (1-100)": "score"})[["annual_income", "score"]]

df.sample(5)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

linkage_data = linkage(df, method='ward', metric='euclidean')
dendrogram(linkage_data)
plt.show()
```

```python
plt.scatter(df["annual_income"], df["score"])
```

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters=5, metric="euclidean", linkage="ward")

labels = model.fit_predict(df)
```

```python
plt.scatter(df["annual_income"], df["score"], c=labels)
```
